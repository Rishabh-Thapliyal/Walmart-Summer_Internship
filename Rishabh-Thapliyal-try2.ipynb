# Autocorrect library

```python
from autocorrect import Speller
```

```python
spell = Speller()

for word in vect.get_feature_names():
    
    if word != spell(word):
        # Get the one `most likely` answer
        print(word, spell(word), sep='-')
```

# TextBlob

```python
from textblob import Word
```

```python
for w in vect.get_feature_names():
    
    if (Word(w).spellcheck()[0][1] > 0.9 or Word(w).spellcheck()[0][1] == 0) and (w != Word(w).spellcheck()[0][0]):
        print(w, Word(w).spellcheck()[0][0], sep='-')
```

# Porter Stemming

```python
# import these modules
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
  
ps = PorterStemmer()
 
for w in vect.get_feature_names():
    
    if w != ps.stem(w):
        print(w, " : ", ps.stem(w))
```

# Snowball Stemmer

```python
import nltk
from nltk.stem.snowball import SnowballStemmer
  
#the stemmer requires a language parameter
snow_stemmer = SnowballStemmer(language='english')

 
for w in vect.get_feature_names():
    
    if w != snow_stemmer.stem(w):
        print(w, " : ", snow_stemmer.stem(w))
```
